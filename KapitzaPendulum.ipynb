### Введение

Давайте рассмотрим [математический маятник](https://ru.wikipedia.org/wiki/%D0%9C%D0%B0%D1%82%D0%B5%D0%BC%D0%B0%D1%82%D0%B8%D1%87%D0%B5%D1%81%D0%BA%D0%B8%D0%B9_%D0%BC%D0%B0%D1%8F%D1%82%D0%BD%D0%B8%D0%BA). Он имеет 2 [положения равновесия](https://ru.wikipedia.org/wiki/%D0%9C%D0%B5%D1%85%D0%B0%D0%BD%D0%B8%D1%87%D0%B5%D1%81%D0%BA%D0%BE%D0%B5_%D1%80%D0%B0%D0%B2%D0%BD%D0%BE%D0%B2%D0%B5%D1%81%D0%B8%D0%B5): "нижнее" (в точке (0, -l)) и "верхнее" (в точке (0, l)), где l - длина маятника, а центр координат расположен в точке крепления маятника (подвеса). Известно, что "нижнее" положение равновесия устойчиво, в то время как "верхнее" - нет. Из-за чего маятник всегда стремится к "нижнему" положению равновесия, за исключением случая, когда он изначально находится в "верхнем", однако любое возмущение выводит маятник из "верхнего" положения равновесия.  

Если же модернизировать маятник и заставить вибрировать точку подвеса (быстро перемещать подвес вверх вниз), то окажется, что "верхнее" положение равновесия также может быть становится устойчивым в зависимости от некоторых условий. Такая система называется Маятником Капицы в честь советского академика [П.Л. Капицы](https://ru.wikipedia.org/wiki/%D0%9A%D0%B0%D0%BF%D0%B8%D1%86%D0%B0,_%D0%9F%D1%91%D1%82%D1%80_%D0%9B%D0%B5%D0%BE%D0%BD%D0%B8%D0%B4%D0%BE%D0%B2%D0%B8%D1%87), который построил теорию для описания данной системы в 1951 году. Капица эксперементально исследовал данную систему, разделяя движение маятника на "быструю" и "медленную" составляющую, а также ввел понятие Эффективного потенциала.

### Обозначения

* $m$ - масса грузика (маятника)
* $v$ - частота гармонических колебаний подвеса
* $a$ - амплитуда колебаний подвеса
* $l$ - длина маятника
* $y_п$ - координата точки подвеса по Оси Ординат
* $\varphi$ - угол между маятником и Осью Ординат

### Уравнение движения

Перейдем в систему отсчета точки подвеса. Эта система будет неинерциальна, и на маятник будет дейсвовать вертикальная сила инерции 

$
F(t) = - m \frac{d^2 y_п} {d^2 t} = mav^2 cos(v t) \\
$ 

Зная чему равна сила инерции, можно найти уравнение движения маятника


<img src="https://github.com/57Ark/Kapitza-Pendulum/blob/1e57940fa8d78aa2788b71b7b13b1a579f9c98da/data/geogebra-export-1.png?raw=true" style="width: 800px">

Из закона сохранения момента импульса 

$
(F - mg)lsin\varphi = m l^2 \frac{d^2 \varphi} {d^2 t}
$

$
\Downarrow \\
$

$
\ddot{\varphi} = \frac{(F - mg)lsin\varphi}{m l^2} \\
$ 

$
\Downarrow \\
$ \
$
\ddot{\varphi} = \frac{(mav^2 cos(v t) - mg)sin\varphi}{m l} \\
$ \
$
\Downarrow \\
$ \
$
\ddot{\varphi} = \frac{(av^2 cos(v t) - g)sin\varphi}{l} \\
$

На самом деле сила $F$ может быть не с плюсом, а с минусом. Это зависит от того, что происходит с точкой подвеса в начальный момент времени (она толкает маятник вверх или тащит его вниз). Так что справедливо также и такое уравнение:  

$
\ddot{\varphi} = \frac{- (av^2 cos(v t) + g)sin\varphi}{l}
$

Стоит заметить, что такое уравнение аналитически не решается, и может быть решено аналитически только с некоторыми допущениями.

### Координаты маятника и построение модели

Нетрудно заметить, что координаты маятника выражаются через $\varphi$ следующим образом:

$
x(t) = l sin(\varphi) \\
$ \
$
y(t) = -l cos(\varphi) \pm a cos(v t) \\
$

Ааналогично уравнению на $\ddot{\varphi}$, в уравнении координаты $y(t)$ слогаемое $a cos(v t)$ может входит как с плюсом, так и с минусом)

Также можно найти координаты точки подвеса:

$
x(t) = 0 \\
y(t) = \pm a cos(vt) \\  
(\textrm{знак должен совпадать со знаком в уравнении координат маятника}) \\
$


Зная координаты маятника и точки подвеса, можно построить анимацию движения маятника, которая будет зависить от $a$, $v$, $l$, $g$ и начального угла $\varphi_0$.

### Быстрое и медленное состовляющие движения

Рассмотрев $l \gg a$ и $ av^2 \gg g$, можно найти уравнение $\beta$ для произвольного значения $\gamma$.


Поскольку $\beta$ - быстрая компонента (то есть со временем она изменяется сильно быстрее, чем $\gamma$), можно предположить, что \

$
\frac{d^2 \beta}{d^2 t} = \frac{d^2 \varphi}{d^2 t} \\
$ \ 
$
\Downarrow \\
$ \
$
\ddot{\beta} = \frac{(av^2 cos(v t) - g)sin(\beta + \gamma)}{l} \quad  (\textrm{т.к. } \varphi = \beta + \gamma) \\
$ \
$
\Downarrow \\
$ \
$
\ddot{\beta} = \frac{av^2 cos(v t) sin(\gamma)}{l} \quad  (\textrm{т.к. } av^2 \gg g \textrm{ и } \gamma \gg \beta) \\
$ \
$
\Downarrow \\
$ \
$
\dot{x} = \frac{av^2 cos(v t) sin(\gamma)}{l} \quad \wedge \\
$ \
$
\dot{\beta} = x \\
$ \
$
\Downarrow \\
$ \
$
\int{l dx} = \int{av^2 cos(v t) sin(\gamma) dt} \quad \wedge \\
$ \
$
\int{1 d\beta} = \int {x dt}  \\
$ \
$
\Downarrow \\
$ \
$
lx = av sin(v t) sin(\gamma) \quad \wedge \\
$ \
$
\beta = \int {x dt} \\
$ \
$
\Downarrow \\
$ \
$
\beta = \int { \frac{av}{l} sin(v t) sin(\gamma) dt} \\
$ \
$
\Downarrow \\
$ \
$
\beta = - \frac{a sin(\gamma) cos(vt)}{l} \\
$


Рассмотрев также большой промежуток времени можно найти $\ddot{\gamma}$ усредняя по большому количеству быстрых колебаний. 


$
\ddot{\gamma} = \frac{\int^t_0{\ddot{\varphi} dt}}{t} \\
$ \ 
$
\Downarrow \\
$ \
$
\ddot{\gamma} = \frac{1}{t}\int^t_0{\frac{(av^2 cos(v t) - g) (sin(\gamma) cos(\beta) + cos(\gamma) sin(\beta))}{l} dt} \quad  (\textrm{т.к. } sin(\varphi) = sin
(\beta + \gamma) = sin(\gamma) cos(\beta) + cos(\gamma) sin(\beta)) \\
$ \
$
\Downarrow \\
$ \
$
\ddot{\gamma} = \frac{1}{lt}\int^t_0{(av^2 cos(v t) - g) (sin(\gamma) cos(\beta) + cos(\gamma) sin(\beta)) dt} \\
$ \

Усредняя, можно предположить $sin(\beta) = \beta$ и $cos(\beta) = 1 - \frac{\beta^2}{2}$ \

$
\ddot{\gamma} = \frac{1}{lt}\int^t_0{(av^2 cos(v t) - g) ((1 - \frac{\beta^2}{2}) sin(\gamma)  + \beta cos(\gamma) ) dt} \\
$ \
$
\Downarrow \\
$ \
$
\ddot{\gamma} = \frac{1}{lt}\int^t_0{(av^2 cos(v t) - g) ((1 - \frac{a^2 sin^2(\gamma) cos^2(vt)}{2 l^2}) sin(\gamma)  - \frac{a sin(\gamma) cos(vt)}{l} cos(\gamma) ) dt}  \\
$ \
$
\Downarrow \\
$ \
$
\ddot{\gamma} = \frac{1}{lt}\int^t_0{(av^2 cos(v t) - g) ((1 - \frac{a^2 sin^2(\gamma) cos^2(vt)}{2 l^2}) sin(\gamma)  - \frac{a sin(\gamma) cos(vt)}{l} cos(\gamma) ) dt} \\
$ \
$
\Downarrow \\
$ \
$
\ddot{\gamma} = \frac{1}{lt}\int^t_0{(av^2 cos(v t) - g) (sin(\gamma) - \frac{a^2sin^3(\gamma)cos^2(vt) + l a sin(2\gamma) cos(vt)}{2l^2}) dt} \\
$ \
$
\Downarrow \\
$ \
$
\ddot{\gamma} = -\frac{1}{lt}\int^t_0{(gsin(\gamma) - cos(vt) \frac{g a sin(2 \gamma) + 2alv^2sin(\gamma)}{2l} - cos^2(vt) \frac{ga^2sin^3(\gamma) - a^2v^2l sin(2\gamma)}{2l^2} + cos^3(vt) \frac{a^3v^2 sin^3(\gamma)}{2l^2})  dt} \\
$ \

Так как рассматривается усреднение и большой промежуток времени, то можно значения интегралов тригономентрических функций в нечетных степенях можно счиать равными нулю. \

$
\ddot{\gamma} = -\frac{1}{lt}\int^t_0{(gsin(\gamma)  - cos^2(vt) \frac{ga^2sin^3(\gamma) - a^2v^2l sin(2\gamma)}{2l^2}) dt} \\
$ \
$
\Downarrow \\
$ \
$
\ddot{\gamma} = -\frac{1}{lt}\int^t_0{(gsin(\gamma) + cos^2(vt) \frac{a^2v^2 sin(2\gamma)}{2l}) dt} \quad  (\textrm{т.к. } l \gg a) \\
$ \
$
\Downarrow \\
$ \
$
\ddot{\gamma} = -\frac{1}{lt}\int^t_0{(gsin(\gamma)  + (cos(2vt) + 1) \frac{a^2v^2 sin(2\gamma)}{4l}) dt} \\
$ \
$
\Downarrow \\
$ \
$
\ddot{\gamma} = -\frac{a^2v^2 sin(2\gamma) + 4l gsin(\gamma)}{4l^2}
$

### Эффективный потенциал

Зная усредненное $\ddot{\gamma}$, можно посчитать действующий на маятник усредненный момент сил относительно точки подвеса. \

$
\textrm{Момент сил  }  M = ml^2 \ddot{\varphi} \\
$  \
$
\Downarrow \\
$ \
$
\textrm{Усредненный момент сил  } M_0 = ml^2 \ddot{\gamma} = -m (\frac{a^2v^2}{4}sin(2\gamma) + l gsin(\gamma)) \\
$

Капица дал определение эффективному потенциалу -- скалярной функции $V(\gamma)$, производная которой по $\gamma$ равна уcредненному моменту сил со знаком -. \

$
V(\gamma)= -\int{M_0 d \gamma} \\
$ \
$
\Downarrow \\
$ \
$
V(\gamma) = \int{m (\frac{a^2v^2}{4}sin(2\gamma) + l gsin(\gamma)) d \gamma} \\
$ \
$
\Downarrow \\
$ \
$
V(\gamma) = \frac{ma^2v^2}{4}\int{sin(2\gamma) d\gamma} + mgl\int{sin(\gamma) d\gamma} \\
$ \
$
\Downarrow \\
$ \
$
V(\gamma) = -\frac{ma^2v^2}{8}cos(2\gamma) - mglcos(\gamma) + C \\
$ \
$
\Downarrow \\
$ \
$
V(\gamma) = -mglcos(\gamma) - \frac{ma^2v^2cos^2(\gamma)}{4} + C \\
$ 

Эффективный потенциал позволяет находить точки равновесия системы. Точка $\gamma_0$ является точкой равновесия, тогда и только тогда, когда $M_0(\gamma_0) = 0$ и $\frac{d M_0}{d\gamma} < 0$ в окрестности $\gamma_0$. То есть $\dot{V}(\gamma_0) = 0$ и $\ddot{V}(\gamma) > 0$ в окрестности $\gamma_0$. Нетрудно заметить, что все $\gamma$, которые удовлетворяют этим условиям, являются точками локального минимума $V(\gamma)$

### Положения устойчивого равновесия

Заметим, что уравнение $V(\gamma)$ - парабола относительно $cos(\gamma)$. Вершина данной параболы  $cos(\gamma_в) = -\frac{4mgl}{2ma^2v^2} = -\frac{2gl}{a^2v^2}$

$cos(\gamma)$ при $\gamma \in$[$0; 2\pi$) может принимать определенное значение 0, 1 или 2 раза, что соответствует количеству положений устойчивого равновесия.
Таким образом:
1) Существует 2 точки равновесия при $1 > cos(\gamma_в) > -1 \Rightarrow av > \sqrt{2gl}$


1. $cos(\gamma_в) > -1  \Rightarrow -\frac{2gl}{a^2v^2} > -1 \quad (\textrm{т.к. } cos(\gamma_в) = -\frac{2gl}{a^2v^2}) \Rightarrow a^2v^2 > 2gl \Rightarrow av > \sqrt{2gl}$


2. $1 > cos(\gamma_в) \Rightarrow 1 > -\frac{2gl}{a^2v^2} \textrm{  - выполняется всегда } $

2) Существует 1 точка равновесия при $cos(\gamma_в) = \pm 1 \Rightarrow av = \sqrt{2gl}$


1. $cos(\gamma_в) = -1  \Rightarrow -\frac{2gl}{a^2v^2} = -1 \Rightarrow a^2v^2 > 2gl \Rightarrow av = \sqrt{2gl}$


2. $cos(\gamma_в) = 1 \Rightarrow -\frac{2gl}{a^2v^2} = 1 \textrm{  - неверно } $


2) Не существует точек равновесия при $|cos(\gamma_в)| > 1 \Rightarrow av < \sqrt{2gl}$


1. $cos(\gamma_в) < -1  \Rightarrow -\frac{2gl}{a^2v^2} < -1 \Rightarrow a^2v^2 < 2gl \Rightarrow av < \sqrt{2gl}$


2. $cos(\gamma_в) > 1 \Rightarrow -\frac{2gl}{a^2v^2} > 1 \textrm{  - неверно } $



Однако, необходимо вспомнить, что изначально было сделано допущение, что $l \gg a$ и $av^2 \gg g$. Данные условия противорячат условию существования ровно одной точки равновесия и условию их несуществования неверны.

Но условие на существование двух точек равновесия по прежнему верно, а значит можно утверждать, что у маятника Капицы существует 2 устойчивых положения равновесия при $av = \sqrt{2gl}$. 

Далее необходимо доказать, что при данных условиях этими положениями является $\gamma = 0$ и $\gamma = \pi$.

1. $\dot{V}(\gamma) = mglsin(\gamma) + \frac{ma^2v^2 sin(\gamma) cos(\gamma)}{2} = mglsin(\gamma) + \frac{ma^2v^2 sin(2\gamma)}{4}$
    1. $\dot{V}(0) = 0$
    2. $\dot{V}(\pi) = 0$
2. $\ddot{V}(\gamma) = mglcos(\gamma) + \frac{ma^2v^2 cos(2\gamma)}{2}$
    1. $\ddot{V}(\gamma) = mgl x + \frac{ma^2v^2 x}{2} > 0$; $\quad (\gamma \in \mathring{U}_\varepsilon(0)$, $x \in \mathring{U}_\varepsilon(1))$
    2. $\ddot{V}(\gamma) = - mgl x + \frac{ma^2v^2 x}{2} > 0 $, $\quad$т.к. $a^2v^2 > 2gl$; $\quad \gamma \in \mathring{U}_\varepsilon(\pi)$,  $x \in \mathring{U}_\varepsilon(1)$
    
Итак, при $av > \sqrt{2gl}$ $0$ и $\pi$ являются положениями устойчивого равновесия
