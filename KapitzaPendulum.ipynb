### Введение

Давайте рассмотрим математический маятник. Он имеет 2 положения равновесия: "нижнее" (в точке (0, -l)) и "верхнее" (в точке (0, l)), где l - длина маятника, а центр координат расположен в точке крепления маятника (подвеса). Известно, что "нижнее" положение равновесия устойчиво, в то время как "верхнее" - нет. Из-за чего маятник всегда стремится к "нижнему" положению равновесия, за исключением случая, когда он изначально находится в "верхнем", однако любое возмущение выводит маятник из "верхнего" положения равновесия.  

Если же модернизировать маятник и заставить вибрировать точку подвеса (быстро перемещать подвес вверх вниз), то окажется, что "верхнее" положение равновесия также может быть становится устойчивым в зависимости от некоторых условий. Такая система называется Маятником Капицы в честь советского академика П.Л. Капицы, который построил теорию для описания данной системы в 1951 году. Капица эксперементально исследовал данную систему, разделяя движение маятника на "быструю" и "медленную" составляющую, а также ввел понятие Эффективного потенциала.

### Уравнение движения

Перейдем в систему отсчета точки подвеса. Эта система будет неинерциальна, и на маятник будет дейсвовать вертикальная сила инерции 
$
F = - m \frac{d^2 y_п} {d^2 t} = mav^2 cos(v t)
$ 
где $y_п$ - координата подвеса по $y$, $a$ - амплитуда колебания подвеса, $v$ - частота колебания подвеса, $m$ - масса маятника, $t$ - время.

Зная чему равна сила инерции, можно найти уравнение движения маятника

<details>
  <summary>Нажмите, чтобы увидеть доказательство</summary>

Из закона сохранения момента импульса
$
(F - mg)lsin\varphi = m l^2 \frac{d^2 \varphi} {d^2 t} 
$ 
$
\Downarrow
$
$
\ddot{\varphi} = \frac{(F - mg)lsin\varphi}{m l^2}
$
$
\Downarrow
$
$
\ddot{\varphi} = \frac{(mav^2 cos(v t) - mg)sin\varphi}{m l}
$
$
\Downarrow
$

</details>

$
\ddot{\varphi} = \frac{(av^2 cos(v t) - g)sin\varphi}{l}
$

На самом деле сила $F$ может быть не с плюсом, а с минусом. Это зависит от того, что происходит с точкой подвеса в начальный момент времени (она толкает маятник вверх или тащит его вниз). Так что справедливо также и такое уравнение:
$
\ddot{\varphi} = \frac{- (av^2 cos(v t) + g)sin\varphi}{l}
$

Стоит заметить, что такое уравнение аналитически не решается, и может быть решено аналитически только с некоторыми допущениями.

### Быстрое и медленное состовляющие движения

$

\Downarrow

$

\Downarrow

$
\dot{x} = \frac{av^2 cos(v t) sin(\gamma)}{l} \quad \wedge \\
\dot{\beta} = x 
$
$
\Downarrow
$
$
\int{l dx} = \int{av^2 cos(v t) sin(\gamma) dt} \quad \wedge \\
\int{1 d\beta} = \int {x dt}

$
\Downarrow
$
$
lx = av sin(v t) sin(\gamma) \quad \wedge \\
\beta = \int {x dt} 
$
$
\Downarrow
$
$
\beta = \int { \frac{av}{l} sin(v t) sin(\gamma) dt}
$
$
\Downarrow
$

$
\beta = - \frac{a sin(\gamma) cos(vt)}{l}

### Эффективный потенциал

Зная усредненное $\ddot{\gamma}$, можно посчитать действующий на маятник усредненный момент сил относительно точки подвеса.

$
\textrm{Момент сил  }  M = ml^2 \ddot{\varphi}
$
$
\Downarrow
$
$
\textrm{Усредненный момент сил  } M_0 = ml^2 \ddot{\gamma} = -m (\frac{a^2v^2}{4}sin(2\gamma) + l gsin(\gamma))
$

Капица дал определение эффективному потенциалу -- скалярной функции $V(\gamma)$, производная которой по $\gamma$ равна уcредненному моменту сил со знаком -.
$
V(\gamma)= -\int{M_0 d \gamma}
$
<details>
  <summary>Нажмите, чтобы увидеть доказательство</summary>

$
\Downarrow
$
$
V(\gamma) = \int{m (\frac{a^2v^2}{4}sin(2\gamma) + l gsin(\gamma)) d \gamma}
$
$
\Downarrow
$
$
V(\gamma) = \frac{ma^2v^2}{4}\int{sin(2\gamma) d\gamma} + mgl\int{sin(\gamma) d\gamma} 
$
$
\Downarrow
$
$
V(\gamma) = -\frac{ma^2v^2}{8}cos(2\gamma) - mglcos(\gamma) + C
$
$
\Downarrow
$
</details>

$
V(\gamma) = -mglcos(\gamma) - \frac{ma^2v^2cos^2(\gamma)}{4} + C
$

Эффективный потенциал позволяет находить точки равновесия системы. Точка $\gamma_0$ является точкой равновесия, тогда и только тогда, когда $M_0(\gamma_0) = 0$ и $\frac{d M_0}{d\gamma} < 0$ в окрестности $\gamma_0$. Нетрудно заметить, что все $\gamma$, которые удовлетворяют этим условиям, являются точками локального минимума $V(\gamma)$
